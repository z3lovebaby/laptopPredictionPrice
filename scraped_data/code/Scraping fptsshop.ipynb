```python
def get_url(product: BeautifulSoup):
    url = 'https://fptshop.com.vn' + product.find('a')['href']
    return url
def cook_soup(url):
    driver.get(url)
    soup_mini = BeautifulSoup(driver.page_source, 'html.parser')
    return soup_mini
```

```python
laptop_1 = soup.find('div', class_ = 'cdt-product prd-lap')
laptop_1_name = laptop_1.find('a', class_ = 'cdt-product__name').text
print('Tên sản phẩm: ',laptop_1_name)
laptop_1_price = laptop_1.find('div', class_ = 'progress pdiscount2').text
print('Giá khuyến mãi: ', laptop_1_price)
laptop_1_origin_price = laptop_1.find('div', class_ = 'strike-price').text 
print('Giá gôc: ',laptop_1_origin_price)
laptop_1_config = laptop_1.find('div', class_ = 'cdt-product__config__param')
laptop_1_att = [] 
for span in laptop_1_config.find_all('span'):
    laptop_1_att.append(span.text)
print(laptop_1_price.strip(' ')[:-1])
print('Màn hình: ', laptop_1_att[0])
print('CPU: ', laptop_1_att[1])
print('RAM: ', laptop_1_att[2])
print('Ổ cứng: ', laptop_1_att[3])
print('Card đồ họa: ', laptop_1_att[4])
print('Trọng lượng (kg): ', laptop_1_att[5])
print(laptop_1_config.find('span')['data-title'])
```

```python
def eat_soup(product: BeautifulSoup):
    cook = cook_soup(get_url(product))
    try: 
        PN = cook.find('div', class_ = 'c-modal__row', id = 'data-13')
        PN = PN.find('ul', class_ = 'st-list')
        PN = PN.li.div.text
    except:
        PN = 'nan'
        
    name = product.find('a', class_ = 'cdt-product__name').text
    price = 'nan'
    price_origin = 'nan'
    try:
        price= product.find('div', class_ = {'progress pdiscount2', 'progress'}).text.split(' ')[0].replace('.','')
        price_origin = product.find('div', class_ = 'strike-price').text.split(' ')[0].replace('.','')
    except AttributeError:
        price= product.find('div', class_ = 'price').text.split(' ')[0].replace('.','')
        price_origin = price
        
    product_config = product.find('div', class_ = 'cdt-product__config__param')
    screen = 'nan'
    RAM = 'nan'
    hard_drive = 'nan'
    graphic_card = 'nan'
    weight = 'nan'
    for span in product_config.find_all('span'):
        if span['data-title'] == 'Màn hình': 
            screen = span.text
        elif span['data-title'] == 'RAM': 
            RAM = span.text
        elif span['data-title'] == 'Card đồ họa': 
            graphic_card = span.text
        elif span['data-title'] == 'Trọng lượng (kg)': 
            weight = span.text
        elif span['data-title'] == 'Ổ cứng': 
            hard_drive = span.text
    
    try:
        CPU_brand = cook.find('div', class_ = 'c-modal__row', id = 'data-0')
        CPU_brand = CPU_brand.find('table', class_ = 'st-table').find_all('tr')
        CPU_name = CPU_brand[0].find_all('td')[1].text + ' ' + CPU_brand[1].find_all('td')[1].text + ' - ' + CPU_brand[2].find_all('td')[1].text

    except:
        CPU_name = 'nan'
    
    try:
        RAM_upgradable = 'No'
        RAM_info = cook.find('div', class_ = 'c-modal__row', id = 'data-1')
        RAM_info = RAM_info.find('table', class_ = 'st-table').find_all('tr')
        if RAM_info[-1].find_all('td')[1] != '--':
            RAM_upgradable = 'Yes'
    except: 
        RAM_upgradable = 'No'

    try:
        battery_wh = cook.find('div', class_ = 'c-modal__row', id = 'data-10')
        battery_wh = battery_wh.find('table', class_ = 'st-table').find_all('tr')[-1].find_all('td')[1].text
    except:
        battery_wh = 'nan'
    
    try:
        storage_upgradable = 'No'
        storage_info = cook.find('div', class_ = 'c-modal__row', id = 'data-4')
        storage_info = storage_info.find('table', class_ = 'st-table').find_all('tr')
        if storage_info[-1].find_all('td')[-1] != '--':
            storage_upgradable = 'Yes'
    except:
        storage_upgradable = 'No'
        
    try:
        screen_info = cook.find('div', class_ = 'c-modal__row', id = 'data-2').find('table', class_ = 'st-table').find_all('tr')
        screen_reso = screen_info[2].find_all('td')[1].text
        screen_rate = screen_info[3].find_all('td')[1].text
        screen_tech = screen_info[6].find_all('td')[1].text
    except:
        screen_reso = 'nan'
        screen_rate = 'nan'
        screen_tech = 'nan'
    
    try:
        dimension = cook.find('div', class_ = 'c-modal__row', id = 'data-12').find('ul', class_ = 'st-list').find_all('li')
        length = dimension[0].div.text
        width = dimension[2].div.text
        height = dimension[4].div.text
        surface = dimension[1].div.text
    except:
        length = 'nan'
        width = 'nan'
        height = 'nan'
        surface = 'nan'
    
    try:
        release = cook.find('div', class_ = 'c-modal__row', id = 'data-13').find('ul', class_ = 'st-list').find_all('li')
        release_year = release[1].div.text
    except:     
        release_year = 'nan'
    
    try:
        ports = [0,0,0,0,0,0]
        ports_info = cook.find('div', class_ = 'c-modal__row', id = 'data-5').find('div', class_='st-modal-tags').find('div', class_ = 'e-tags').find_all('a')
        try:
            for a in ports_info:
                if 'hdmi' in a.text.lower():
                    ports[0] += int(a.text[0])
                elif 'jack' in a.text.lower():
                    ports[1] += int(a.text[0])
                elif 'lan' in a.text.lower():
                    ports[2] += int(a.text[0])
                elif 'type-c' in a.text.lower():
                    ports[3] += int(a.text[0])
                elif 'usb' in a.text.lower():
                    ports[4] += int(a.text[0])
                elif 'sd' in a.text.lower():
                    ports[5] += int(a.text[-1][0])
        except:
            pass
    except:
        ports = ['nan', 'nan', 'nan', 'nan', 'nan', 'nan']
    
    try:
        keyboard_light = cook.find('div', class_ = 'c-modal__row', id = 'data-9').find('table', class_ = 'st-table').find_all('tr')[2].find_all('td')[1]
        if keyboard_light.text != '--':
            keyboard_light = 'Yes'
        else: keyboard_light = 'No'
    except:
        keyboard_light = 'No'
        
    try:
        os = cook.find('div', class_ = 'c-modal__row', id = 'data-11').find('table', class_ = 'st-table').find_all('tr')[0].find_all('td')[1].text
    except:
        os = 'nan'
    
    try:
        rating = int(cook.find('div', class_ = 'point').text[0])
    except:
        rating = 'nan'
    rating_detail = []
    
    try: 
        list = cook.find('div', class_="c-progress-list").find_all('div', class_ = 'c-progress-item')
        for item in list:
            rating_detail.append(item.find_all('span')[1].text)
    except: 
        rating_detail = ['nan', 'nan', 'nan', 'nan', 'nan']
    
    try: 
        rating_count = cook.find('div', class_ = 'c-rate__left text-center').find('span').text.split(' ')[0]
        comment_count  = cook.find_all('h2', class_="card-title card-title--badge")[1].find('span').text
    except: 
        rating_count = 'nan'
        comment_count = 'nan'
        
    info = [PN, rating,  rating_count,  comment_count, name, price, CPU_name, RAM, RAM_upgradable, graphic_card, battery_wh,
           hard_drive, storage_upgradable, screen, screen_reso, screen_rate, screen_tech, weight, length, width, height, os,
           surface, release_year, ports, keyboard_light]
    return info
```

```python
fpt_url = 'https://fptshop.com.vn/may-tinh-xach-tay?sort=ban-chay-nhat&trang=6'
options = EdgeOptions()
options.use_chromium = True
driver = Edge(options = options)
driver.get(fpt_url)
soup = BeautifulSoup(driver.page_source, 'html.parser')
all_lap = soup.find_all('div', class_ = {'cdt-product prd-lap', 'cdt-product prd-lap product-sale'})
records = []
for meal in all_lap:
    records.append(eat_soup(meal))
with open('result_detail_final.csv','w',newline='',encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['PN','rating','rating_count','name','price','CPU', 'RAM', 'RAM_upgradable', 'GPU', 
                     'battery_wh', 'hard_drive', 'storage_upgradable', 'screen_size', 
                     'display_resolution', 'display_rating', 'screen_technology', 'weight',
                    'length', 'width', 'height', 'OS', 'surface_material', 'release_year', 'ports', 'keyboard_light'])
    writer.writerows(records)
```
