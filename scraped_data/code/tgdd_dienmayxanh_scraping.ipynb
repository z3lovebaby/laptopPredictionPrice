```python
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
import selenium.webdriver.support.expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd

driver = webdriver.Chrome('/home/hung/Downloads/chromedriver')
```

# Prototype

```python
spec_id_to_name = {
    '28727':'cpu_name',
    '28859':'cpu_ncores',
    '28860':'cpu_nthreads',
    '93':'cpu_basefreq',
    '97':'cpu_turbofreq',
    '28861':'cpu_cache',
    
    '146':'ram_size',
    '149':'ram_type',
    '155':'ram_clock',
    '137':'ram_upgradable',
    
    '184':'storage',
    '28123':'gpu',
    '228':'battery',
    
    '187':'displ_size',
    '189':'displ_resolution',
    '29056':'displ_rate',
    '186':'displ_tech',
    '480':'displ_touchscreen',

    '196':'sound',
    '200':'ports',
    '218':'sd_card',
    '206':'wless_connect',
    '223':'webcam',
    '201':'special',
    '10741':'backlight_key',

    '7779':'weight_dims',
    '7903':'surface',
    '8599':'os',
    '22711':'launch'
}
```

```python
def specs_scrap(data_id,soup0):
    '''data_id: str'''
    li = soup0.find('li',{'data-id':data_id})
    if not li:
        return ''
    ls = []
    for p in li.find_all('p',{'class':'circle'}):
        if p.text and p.text not in ls: ls.append(p.text.strip())
#         else: ls.append(p.a.text.strip())

    for a in li.find_all('a'):
        if a.text and a.text not in ls: ls.append(a.text.strip())
        
    for span in li.find_all('span'):
        if span.text and span.text not in ls: ls.append(span.text.strip())

#     if not os:
#         try:
#             os = li.span.text.strip()
#         except:
#             os = '|'.join(a.text.strip() for a in li.find_all('a'))
    return "|".join(ls)
```

```python
def product_scrap(url0,url1):
    """
    Scrap all attributes from a laptop
    input 2 laptop urls from thegioididong.com and dienmayxanh.com, these 2 urls must have the same product extension
    Return: a list of values of all attributes from laptop 
    """
    driver.get(url0)
    soup0 = BeautifulSoup(driver.page_source, 'html.parser')
    
    #start scraping title, price, rating, review_count, cmt_count
    try: title = soup0.find('h1').text.strip()
    except: return ['']*(6+len(spec_id_to_name))
    try: price = soup0.find('p',{'class':'box-price-present'}).text.strip()
    except: price = ''
    try: rating = soup0.find('p',{'class':'point'}).text.strip()
    except: rating = ''
    try: review_count = soup0.find('a',{'class':'rating-total'}).text.strip()
    except: review_count = ''
    try: cmt_count = soup0.find('span',{'class':'totalcomment'}).text[:22].strip()
    except: cmt_count = ''

    #click the button
    try:
        btn = driver.find_element(By.CSS_SELECTOR,'.btn-detail.btn-short-spec').click()
        WebDriverWait(driver,timeout=5).until(EC.visibility_of_element_located((By.CSS_SELECTOR,'.parameter-all')))
    
        #start scraping specs
        soup0 = BeautifulSoup(driver.page_source, 'html.parser')
        specs_row = [specs_scrap(id,soup0) for id in spec_id_to_name]
    except: specs_row = ['']*len(spec_id_to_name)
    #get comment numbers from dienmayxanh
    driver.get(url1)
    soup1 = BeautifulSoup(driver.page_source, 'html.parser')
    try: cmt_count1 = soup1.find('span',{'class':'totalcomment'}).text[:22].strip()
    except: cmt_count1 = ''
    cmt = cmt_count + '|' + cmt_count1
    return [title,price,rating,review_count,cmt] + specs_row  + [url0]
```

```python
product_scrap('https://www.thegioididong.com/laptop/asus-vivobook-x515ma-n4020-br480w?src=osp#top-tskt', 
              'https://www.dienmayxanh.com/laptop/asus-vivobook-x515ma-n4020-br480w?src=osp#top-tskt')
```

# Main function

```python
def main():
    headers = ['title','price','rating','review_count','cmt_count'] + list(spec_id_to_name.values()) + ['url']
    result = pd.DataFrame(columns = headers)
    scraped = 0 #track the number of scraped laptop
    
    url = "https://www.thegioididong.com/laptop#pi=10"
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    laptops = soup.find_all('li',{'class':'item ajaxed __cate_44'})

    for item in laptops:
        url0 = 'https://www.thegioididong.com' + item.a.get('href') + '#top-tskt'
        url1 = 'https://www.dienmayxanh.com' + item.a.get('href') + '#top-tskt'
        print(url0)
        #scrap the product and append to dataframe
        result.loc[len(result.index)] = product_scrap(url0, url1)

        scraped += 1
        print(len(laptops),"scrapped:",scraped)
    return result
```

```python
pd.set_option('display.max_columns', 34)
result = main()
```
