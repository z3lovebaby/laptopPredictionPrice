# Scrap CellphoneS

```python
def get_details(url):
    """Extract the details of an item.
    
    Argument:
        url -- url to the item's page.
        
    Return:
        a dictionary containing all details scrapped for the specified item.
    """
    
    data = {}
        
    driver.get(url)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    data['url'] = url
    
    # name
    data['item_name'] = soup.find('div', {'class': 'box-name__box-product-name'}).h1.text.strip()

    # price
    price_box = soup.find('div', 'box-info__box-price')
    try:
        data['special_price'] = price_box.find('p', {'class': 'special-price'}).text[:-2]
    except:
        data['special_price'] = None
    try:
        data['old_price'] = price_box.find('p', {'class': 'old-price'}).text[:-2]
    except:
        data['old_price'] = None
    
    # rating
    rating_raw = soup.find_all('div', 'item-statistical')
    rating = {
        (level.find('p', 'number-star').strong.text + 'star'): level.find('p', 'number-percent').text[:-9]
        for level in rating_raw
    }
    data.update(rating)
    
    # specifications
    info_table = soup.find('div', {'id': 'technicalInfoModal'}).find_all('th')

    infos = {
        info_table[2*i].text: info_table[2*i + 1].text for i in range(len(info_table)//2)
    }
    data.update(infos)
    
    # comment count
    WebDriverWait(driver,timeout=5).until(EC.text_to_be_present_in_element((By.ID,'total_comment'),"n)"))
    try:
#         print(soup.find('p', {'id': 'total_comment'}).text)
        data['comment_count'] = driver.find_element(By.ID,'total_comment').text.split()[3][1:]
    except:
        data = get_details(url)
    
    return data
        
```

```python
page_url = "https://cellphones.com.vn/laptop.html"

driver = webdriver.Chrome()

driver.get(page_url)
soup = BeautifulSoup(driver.page_source, 'html.parser')

# click to the "show more" button to get full list of laptops
load = True
while load:
    try:
        WebDriverWait(driver, 60).until(
                EC.visibility_of_element_located((By.CLASS_NAME, 'btn-show-more'))
            ).click()        
    except:
        load = False
```

```python
driver = webdriver.Chrome()
records = []
for url in url_list:
    record = get_details(url)
    records.append(record)
    print(record['item_name'])
```

# Scrap LaptopArena

```python
def scrap_brand(brand_url):

    driver.get(brand_url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    if len(soup.find_all('div', {'class': "display_btn_container hide-for-small-only"})) != 0:
        scrap_series(brand_url)
    else:

        family_items = soup.find_all('div', {'class': "product"})

        family_urls = [brand_url + '/' + '%20'.join(item.text.strip().lower().split()) for item in family_items]

        for url in family_urls:        ########################################
            print("\t ***************************** FAMILY *****************************")
            print("\t", url)
            print("\t ******************************************************************")

            scrap_family(url)
            

def scrap_family(family_url):
    
    driver.get(family_url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    if len(soup.find_all('div', {'class': "display_btn_container hide-for-small-only"})) != 0:
        scrap_series(family_url)
    else:

        series_items = soup.find_all('div', {'class': "product"})

        series_urls = [family_url + '/' + '%20'.join(item.text.strip().lower().split()) for item in series_items]

        for url in series_urls:         ########################
            print("\t\t ----------- SERIES ----------")
            print("\t\t", url)
            print("\t\t -----------------------------")

            scrap_series(url)
            

def scrap_series(series_url):
    
    driver.get(series_url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # click to the "Load more laptops" button to get full list of laptops
    print("\t\tClicking")
    for _ in range(8):               #############################
        try:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            driver.find_element(By.CSS_SELECTOR, 'span.button.load_more_products').click()    # must replace [space] with [.] 'cause [space] separates into three classes  
            time.sleep(2)
        except:
            pass
            

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    product_items = soup.find_all('div', {'class': "product"})

    product_urls = ['https://www.laptoparena.net' + item.a.get('href') for item in product_items]

    for url in product_urls:     ###########################
        print("\t\t\t", url)
        scrap_product(url)
        

def scrap_product(product_url):

        driver.get(product_url)
        soup = BeautifulSoup(driver.page_source, 'lxml')

        info_raw = soup.find_all('td', {'class': "table_cat"})
        
        info = {item.text: item.nextSibling.text for item in info_raw}
        info['product_url'] = product_url
        info['product_title'] = soup.find('h1', {'class': "laptop_h1"}).text

        records.append(info)
        
def scrap_product_by_ID(product_id):
    
        product_url = 'https://www.laptoparena.net/product/' + str(product_id)

        scrap_product(product_url)
```
