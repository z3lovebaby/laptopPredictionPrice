```python
import pandas as pd
```

```python
data = pd.read_csv('./result_detail_final.csv', index_col = False)
data.head()
```

```python
data[['Brand', 'Series']] = data.name.str.split(n = 1, expand=True)
data = data.drop('name', 1)
data.head()
```

```python
data['RAM'] = data['RAM'].str.split(' ', 1).str[0]
data.head()
```

```python
data[['storage_type', 'storage_size']] = data['hard_drive'].str.split(' ', n = 1, expand = True)
data = data.drop('hard_drive', 1)
data.head()
```

```python
data['storage_size'] = data['storage_size'].str.extract('(\d+)')
data['storage_size']
```

```python
data['screen_size'] = data['screen_size'].str.strip('"')
data['screen_size']
```

```python
data['display_resolution'] = data['display_resolution'].str.split(' ', n = 3).str[0:3].str.join(' ')
data['display_resolution'] 
```

```python
data['display_rating'] = data['display_rating'].str.split(' ', 1).str[0]
data['display_rating'] 
```

```python
data[['cpu_brand', 'cpu_code']] = data['CPU'].str.split('-', n = 1, expand = True)
data.head()
```

```python
for row, brand in enumerate(data.loc[:,'cpu_brand']):
    try:
        list = brand.split(' ', 1)
        if 'apple' in brand.lower():
            data.at[row, 'cpu_brand'] = list[0]
            data.at[row, 'cpu_code'] = list[1]
        else: 
            data.at[row, 'cpu_brand'] = list[0]
    except: 
        continue
data['cpu_brand']
```

```python
cols = data.columns.tolist()
print(cols)
```

```python
cols  = ['PN', 'rating', 'rating_count','comment_count' ,'Brand', 'Series','price', 'cpu_brand', 'cpu_code',
         'RAM', 'RAM_upgradable', 'GPU', 'battery_wh', 'storage_type', 'storage_size',
         'storage_upgradable', 'screen_size', 'display_resolution', 'display_rating', 'screen_technology', 
         'weight', 'length', 'width', 'height', 'OS', 'surface_material', 'release_year', 
        'hdmi', 'jack', 'lan', 'type_c', 'usb', 'sd', 'keyboard_light']
data = data.reindex(columns = cols)
data.iloc[0,:]
```
